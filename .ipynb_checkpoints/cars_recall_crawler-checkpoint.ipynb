```python
import urllib2
import bs4
import numpy as np
import pandas as pd
from multiprocessing import Pool
from sqlalchemy import create_engine
import os
import warnings
warnings.filterwarnings(action='ignore')
```

```python
def find_href(url_root,page_number):
    url='https://www.cars.com/news/recalls/?pageIndex=%s'%page_number
    f=urllib2.urlopen(url)
    soup=bs4.BeautifulSoup(f)
    href_list=[]
    href_list_parent=soup.find_all('h4')
    for item in href_list_parent:
        if item['class'][0]=='mobile':
            href=item.a['href']
            href=url_root+href
            tmp=[page_number,href]
            href_list.append(href)
        else:
            pass
    return href_list
```

```python
def find_article(soup):
    art=''
    title=soup.title.string
    article=soup.find_all('section')[0]
    for item in article.children:
        if item!='\n':
            para=item.string
            if para:
                art=art+para+'\n'
            else:
                pass
    if len(art)==0:
        try:
            item_list=article.find_all('p')
            for item in item_list:
                if item!='\n':
                    para=item.get_text()
                    art=art+para+'\n'
                else:
                    pass
        except:
            title=None
            art=None
    else:
        pass
    
    return title,art
```

```python
def execute(page_number):
    engine=create_engine("mysql+pymysql://liyuefan:1994050306@localhost/gtja_intern?charset=utf8")
    url_root='https://www.cars.com'
    href_list=find_href(url_root,page_number)
    recall_list=[]
    cnt=0
    for url in href_list:
        f=urllib2.urlopen(url)
        soup=bs4.BeautifulSoup(f)
        title,art=find_article(soup)
        if art==None:
            cnt+=1
            print url, 'None', cnt
            art=''
        else:
            print url
        tmp=[page_number,url,title,art]
        recall_list.append(tmp)
    recall_df=pd.DataFrame(recall_list,columns=['page_number','url','title','article'])
#     recall_df.to_sql('cars_recall',engine,index=False,if_exists='append')
    return recall_df
```
