```python
import numpy as np
import pandas as pd
import datetime

def construct(category)
    name=locals()

    df=pd.read_csv('existing_%s.csv'%category,encoding='gbk')
    df['update_date']=pd.to_datetime(df['update_date'])
    df=df[['company_name_2','hold_vol_buy','hold_vol_buy_chg','update_date']]
    df=df.drop_duplicates()
    df=df[df['update_date']>=pd.to_datetime(datetime.date.today())-np.timedelta64(3,'M')]


    df=df.groupby(['company_name_2','update_date']).sum()
    df=df[df['hold_vol_buy']>=10000]
    df.sort(['hold_vol_buy'])

    pool=list(df.index.levels[0][list(df.index.labels[0])])
    pool=pool[:np.min((10,len(pool)))]


    df=pd.read_csv('existing_%s.csv'%category,encoding='gbk')
    df['update_date']=pd.to_datetime(df['update_date'])
    df=df[['company_name_2','hold_vol_buy','hold_vol_buy_chg','update_date']]
    df=df.drop_duplicates()
    df=df[df['update_date']>=pd.to_datetime(datetime.date.today())-np.timedelta64(3,'M')]


    i=0
    total=pd.DataFrame()
    for item in pool:
        print item
        name['df_%s'%i]=df[df['company_name_2']==item]
        name['df_%s'%i]=name['df_%s'%i].groupby(['update_date']).sum()
        i+=1


    total.to_csv('total_duotou_%s.csv'%category,encoding='gbk',index=True)

    df=pd.read_csv('existing_%s.csv'%category,encoding='gbk')
    df['update_date']=pd.to_datetime(df['update_date'])
    df=df[['company_name_3','hold_vol_sell','hold_vol_sell_chg','update_date']]
    df=df.drop_duplicates()
    df=df[df['update_date']>=pd.to_datetime(datetime.date.today())-np.timedelta64(3,'M')]




    df=df.groupby(['company_name_3','update_date']).sum()
    df=df[df['hold_vol_sell']>=10000]
    df.sort(['hold_vol_sell'])

    pool=list(df.index.levels[0][list(df.index.labels[0])])
    pool=pool[:np.min((10,len(pool)))]


    i=0
    total=pd.DataFrame()
    for item in pool:
        print item
        name['df_%s'%i]=df[df['company_name_3']==item]
        name['df_%s'%i]=name['df_%s'%i].groupby(['update_date']).sum()
        i+=1

    total.to_csv('total_kongtou_%s.csv'%category,encoding='gbk',index=True)
```
